```python
import pandas as pd
import numpy as np
```

Load csv files

```python
group1 = pd.read_csv("../data/raw_sample2/FOOD-DATA-GROUP1.csv")
group2 = pd.read_csv("../data/raw_sample2/FOOD-DATA-GROUP2.csv")
group3 = pd.read_csv("../data/raw_sample2/FOOD-DATA-GROUP3.csv")
group4 = pd.read_csv("../data/raw_sample2/FOOD-DATA-GROUP4.csv")
group5 = pd.read_csv("../data/raw_sample2/FOOD-DATA-GROUP5.csv")
```

All five groups have same columns. So, we can simply combine them.

```python
food_df = pd.concat(
    [group1, group2, group3, group4, group5],
    ignore_index=True
)
```

```python
food_df.shape
```

```python
food_df.head()
```

```python
food_df.tail()
```

Check for duplicates

```python
food_df.duplicated().sum()
```

```python
food_df["food"].duplicated().sum()
```

```python
food_df.isnull().sum()
```

```python
food_df.describe()
```

Drop Unnecessary Columns

These are simply index columns that got saved into the CSV.

```python
food_df = food_df.drop(
    columns=[
        "Unnamed: 0",
        "Unnamed: 0.1"
    ]
)
```

```python
food_df.head()
```

Check for duplicates in food

```python
food_df["food"].duplicated().sum()
```

```python
food_df[food_df["food"].duplicated()]
```

```python
food_df["food"].sample(20)
```

```python
food_df.columns = (
    food_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
food_df.columns
```

let's give a final check for duplicates

```python
food_df["food"].duplicated().sum()
```
