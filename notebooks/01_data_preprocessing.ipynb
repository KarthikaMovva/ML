Change the column name "description" in food_category_df. Because, we already have a column with name "description" in food_df. If, we join them directly. It creates "description_x" and "description_y" which is not clear. So, before left join. We will change the column name "description" of food_category_df.

```python
food_category_df = food_category_df.rename(
    columns={
        "description": "category"
    }
)
```

Let's use left join

```python
food_with_category = pd.merge(
    food_df,
    food_category_df,
    left_on="food_category_id",
    right_on="id",
    how="left"
)
```

```python
print(food_df.shape)
print(food_with_category.shape)
```

Let's pick the target nutrients with their id's.

```python
TARGET_NUTRIENTS = {
    "Energy": 1008,
    "Protein": 1003,
    "Fat": 1004,
    "Carbohydrate": 1005,
    "Fiber": 1079,
    "Sugar": 1063,
    "Sodium": 1093,
    "Cholesterol": 1253
}
```

Our goal is to convert this:

fdc_id	nutrient_id	amount

1102647	1008	52

1102647	1003	0.3

1102647	1004	0.2

1102647	1005	13.8

into

fdc_id	Energy	Protein	Fat	Carbohydrate	Fiber	Sugar	Sodium	Cholesterol

1102647	52	0.3	0.2	13.8	2.4	10.4	1	0

This transformation is called pivoting.

```python
selected_nutrients = food_nutrient_df[
    food_nutrient_df["nutrient_id"].isin(TARGET_NUTRIENTS.values())
]
```

Keep only the required rows with target nutrients and delete the rest of the rows and this is called row filtering.

```python
selected_nutrients = selected_nutrients.merge(
    nutrient_df,
    left_on="nutrient_id",
    right_on="id",
    how="left"
)
```

pivot()

Works only if there is exactly one value for each (fdc_id, nutrient) pair.

If the USDA dataset has duplicate measurements for the same nutrient, pivot() throws an error.

pivot_table()

Allows duplicates and lets us decide how to combine them.

```python
nutrition_wide = selected_nutrients.pivot_table(
    index="fdc_id",
    columns="name",
    values="amount",
    aggfunc="mean"
)
```

index="fdc_id"

This becomes the row identifier.

Instead of

Apple
Energy

Apple
Protein

Apple
Fat

we get

once.

columns="name"

Every unique nutrient becomes a new column.

Instead of

Energy

Protein

Fat

as rows,

they become

Energy | Protein | Fat

as columns.

values="amount"

These become the actual values inside the table.

aggfunc="mean"

Suppose a food has two Energy values.

52

The pivot stores

51

using the average.
